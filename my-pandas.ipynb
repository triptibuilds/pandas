# DATASTRUCTURES
1. Series:  key difference between Series and ndarray is that operations between Series automatically align the data based on label. 


```python
import pandas as pd

sList = pd.Series([1,2,3],index = [1,2,3])
sDict = pd.Series({1:'object1',2:'Object2'})
sDictIdx = pd.Series({"a": 0.0, "b": 1.0, "c": 2.0},index = ['b','c','d'])
sScalar = pd.Series(5,index = [1,2,3,4],dtype = 'int64')
sNDarray = pd.Series([[1,2,3,'a'],[4,5,6]],index = ['a','b'])

print(sList)
print(sDict)
print(sDictIdx)
print(sScalar)
print(sNDarray)

print(sNDarray['a'][0])
sList['new'] = 7
print(sList)

print(sList+sScalar)
```
