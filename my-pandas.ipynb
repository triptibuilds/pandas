### *DataFrame*: MERGE

`.concat()`

- objs: list of dfs to merge

- axis: 0 - stack vertically ; 1 - stack horizontally

- join: outer - union of cols ; inner - intersection of cols

- ignore_index: True - reset index ; False - keep og index

- verify_integrity: if duplicates, produce error

- sort: True/False

- copy: True/False

- keys = [] to differentiate b/w dfs

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)

df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)


df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10"],
        "B": ["B8", "B9", "B10"],
        "C": ["C8", "C9", "C10"],
        "D": ["D8", "D9", "D10"],
    },
    index=[8, 9, 10],
)

frames = [df1,df2,df3] # concat makes a full copy of the data, and iteratively reusing it can create unnecessary copies. It is better to gather dfs in a list first
pd.concat(frames,axis=1,keys=['df1','df2','df3'])


```

`merge()`

- df1,df2

- how: 'left/right/inner/outer'

- on: common col name

- left_on & right_on: when col names are diff data is same

- left_index & right_index: True/False

- sort: True

- suffixes = () to identify dataframes when common col names

- indicator: True/False - shows source

- copy: True/False

- validate: 'one_to_one/one_to_many/many_to_one/many_to_many' (error)

```python
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['A', 'B', 'C']
})

df2 = pd.DataFrame({
    'id': [2, 3, 4],
    'marks': [80, 90, 70]
})

print(pd.merge(df1,df2,how='left',on='id'))
print(pd.merge(df1,df2,how='right',on='id'))
print(pd.merge(df1,df2,how='inner',on='id'))
print(pd.merge(df1,df2,how='outer',on='id',indicator=True))
print(pd.merge(df1,df2,how='cross',indicator=True))
```

`.join()` : index(by default)

- other df

- how

- on : if col based join

- lsuffix & rsuffix

- sort

- validate

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2']
}, index=[0, 1, 2])

df2 = pd.DataFrame({
    'B': ['B0', 'B1', 'B2']
}, index=[1, 2, 3])

df1.join(df2, how='outer', lsuffix='_L', rsuffix='_R', sort=True)
```

### *DataFrame* : Grouping

`.groupby()`

- by

- axis

- level: multi-index df

- as_index: True/False

- sort

- group_keys: True/False

- observed

- dropna: True/False

```python
df = pd.DataFrame({
    'Department': ['IT', 'HR', 'IT', 'HR', 'IT'],
    'Employee': ['A', 'B', 'C', 'D', 'E'],
    'Salary': [50000, 40000, 60000, 45000, 70000]
})


df.groupby('Department',as_index=False)['Salary'].sum()
```

```python
df = pd.DataFrame({
    'Dept': ['IT', 'IT', 'HR', 'HR'],
    'Salary': [50000, 60000, 40000, 45000]
})


print(df.groupby('Dept')['Salary'].apply(lambda x: x.max() - x.min()))

df['Range'] = (df.groupby('Dept')['Salary'].transform(lambda x: x.max() - x.min()))

print(df)
```

### *DataFrame* : Reshaping

**Multi Index DataFrame**

```python
# from_tuples

index = pd.MultiIndex.from_tuples([
    ('India',2021),
    ('India',2022),
    ('China',2021),
    ('China',2022)
], names= ['Country', 'Year'])

df = pd.DataFrame({
    'GDP': [2.9,3.1,21.0,22.5]
}, index = index)

print(df)

# from_product
index = pd.MultiIndex.from_product(
    [['India','China'],[2021,2022]],
    names=['Country','Year']
)

df = pd.DataFrame({
    'GDP': [2.9,3.1,21.0,22.5]
}, index = index)

print(df)

# from columns
data = {
    'Country': ['India','India','China','China'],
    'Year': [2020,2021,2020,2021],
    'GDP': [2.9,3.1,21.0,22.5]
}

df = pd.DataFrame(data)
df.set_index(['Country','Year'],inplace = True)
print(df)



```

#### *DATAFRAME*: RESHAPING

**`.stack()`**: compresses a level in dataframe columns

```python
index = pd.MultiIndex.from_product([['India','China'],[2021,2022]],names=['Country','Year'])
df = pd.DataFrame({'GDP':[2.4,2.5,1.8,1.2]},index=index)
df = df.stack()
print("stacked: \n",df)
print("\nunstacked:\n",df.unstack())
```

**`.pivot_table()`**

- df

- values: Which column should I calculate on ?

- index: Which column should become index ?

- columns: Which column values should become NEW columns ?

- aggfunc: How should I combine them ?

- fill_value: missing

- margins: total [t / f]

```python
df = pd.DataFrame({
    'Date':['Jan','Jan','Feb','Feb'],
    'Category':['A','B','A','B'],
    'Sales':[100,150,150,100]
})

pd.pivot_table(df,values='Sales',index='Date',columns='Category',aggfunc='sum',margins=True)
```
