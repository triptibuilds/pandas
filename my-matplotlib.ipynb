```python
import matplotlib.pyplot as plt
```

**`.plot(x,y, parameters)`**

| Parameter | Short Form | Possible Values / Examples |
|---|---|---|
| color | c | `'r'`, `'g'`, `'b'`, `'black'`, `'yellow'`, `'cyan'`, `'magenta'` |
| linewidth | lw | Any number like `1`, `2`, `5` |
| linestyle | ls | `'-'`, `'--'`, `':'`, `'-.'` |
| marker | — | `'o'`, `'*'`, `'+'`, `'x'`, `'s'`, `'^'`, `'D'` |
| markersize | ms | Any number like `5`, `10`, `15` |
| markerfacecolor | mfc | Any valid color |
| markeredgecolor | mec | Any valid color |
| label | — | Any string like `'Sales'`, `'Profit'` |
| alpha | — | `0` to `1` |
| drawstyle | — | `'default'`, `'steps'`, `'steps-pre'`, `'steps-mid'`, `'steps-post'` |
| zorder | — | Any integer like `1`, `5`, `10` |
| visible | — | `True`, `False` |
| antialiased | aa | `True`, `False` |
| dash_capstyle | — | `'butt'`, `'round'`, `'projecting'` |
| solid_capstyle | — | `'butt'`, `'round'`, `'projecting'` |
| animated | — | `True`, `False` |
| picker | — | `True`, `False` |
| scalex | — | `True`, `False` |
| scaley | — | `True`, `False` |

#### Format String Examples

| Format String | Meaning |
|---|---|
| `'ro--'` | Red + Circle + Dashed |
| `'g^:'` | Green + Triangle + Dotted |
| `'bs-'` | Blue + Square + Solid |
| `'mx-.'` | Magenta + X Marker + Dash-dot |


**`legend()`**: to show the labels passed during  plot()

**`show()`**: show the plots

**`title()`**: title of the plot

**`xlabel()`**: labels for x axis

**`ylabel()`**: labels for the y axis

**`xticks(arr)`**: x axis dustribution

**`ticks_params(axis)`**: modify ticks

```python
x = [2023,2024,2025,2026]
y1 = [20,27,25,26]
y2 = [15,5,26,32]
y3 = [17,20,12,23]

plt.title('class size',
          fontsize=16,
          fontweight='bold',
          color='black')

plt.xlabel('year',
           fontsize=10,
           color='#870017',
           fontstyle='italic')

plt.ylabel('students',
           color='#870017',
           fontsize=10,
           fontstyle='italic')

line_style = dict(c='#040169',
                  ls='None',
                  marker='x',
                  ms=10,
                  mfc='black',
                  mec='black',
                  alpha=1,
                  zorder=3)

plt.xticks(x)
plt.tick_params(axis='both',
                colors='blue')

plot1 = plt.plot(x, y1, c='#dea004',lw=2,ls='--',marker='o',ms=7,mfc='white',mec='black',label='sample1',alpha=0.9, drawstyle='default',zorder=1)
plot2 = plt.plot(x,y2,**line_style,label='sample2')
plot3 = plt.plot(x,y3,'k^-.',alpha=0.7,zorder=2,label='sample3',lw=2)

plt.legend()
plt.show()
```
