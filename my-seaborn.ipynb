### SCATTER PLOT

`**set_style**`:

- ticks

- whitegrid 

- dark

- darkgrid

- white

**`despine()`**: remove top and right spine, can remove left and bottom as well ( left = True)

**`set_context()`**:

- paper

- notebook 

- talk

- poster

```python
sns.set_style('ticks')
sns.set_context('notebook')
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = 'sex', palette = 'Dark2',style='sex',size='island', markers = ['o','^'])
sns.despine()

plt.title("Body mass and Flipper length relation w the sex and island", pad=30)

plt.show()
```

### STRIPPLOT

```python
sns.stripplot(data=penguins, x = 'species', y='body_mass_g',hue='sex', dodge=True)
```

### SWARMPLOT

```python
sns.set_context('talk')

sns.swarmplot(data=penguins, x = 'species', y='body_mass_g',hue='island')
sns.despine()

plt.title('Species and bodymass distribution on islands', pad=30)
plt.show()
```

### HISTOGRAM

stat: frequency, count, probabilty, density


```python
sns.set_context('notebook')

sns.histplot(data = penguins, x = 'body_mass_g',bins=15, hue = 'sex',stat='frequency', palette='bright', multiple='stack',kde=True)
sns.despine()

plt.title("Body mass distribution",pad=30)
plt.show()
```

### REGRESSION PLOT

```python
sns.regplot(data = penguins, x= 'body_mass_g', y='flipper_length_mm',
            scatter_kws={'s':20,'edgecolor':'blue','color':'blue','alpha':0.4}, 
            line_kws={'linewidth':3,'color':'red','linestyle':'-'},
            ci=None,order=2)
```

### LINE PLOT

```python
sns.set_style('whitegrid')

sns.lineplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = 'sex',marker='*',errorbar=None)
```

### JOINT PLOT

```python
sns.jointplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g",kind="hex",color='green',height=4)
```

### BAR PLOT

```python
sns.barplot(data = penguins, x = 'species', y = 'body_mass_g', hue = 'sex', estimator = np.var)
```

### COUNT PLOT

```python
sns.countplot(data = penguins, x= 'species', hue='sex')
```

### BOX PLOTS

```python
sns.boxplot(data = penguins, x ='species', y='body_mass_g',hue='island')
```

### VOILEN PLOT

```python
sns.violinplot(data = penguins, x ='species', y='body_mass_g',hue='island',inner='point')
```

```python
fig, axis= plt.subplots(1,2,figsize=(15,5))

sns.boxplot(data = penguins, x ='species', y='body_mass_g',hue='sex',ax=axis[0])
sns.violinplot(data = penguins, x ='species', y='body_mass_g',hue='sex',ax=axis[1],split=True)
```

### KDE PLOT

```python
sns.kdeplot(data = penguins, x = 'body_mass_g',hue='species',fill=True,bw_adjust=3)
```

## HEATMAP

```python
correlation = penguins.corr(numeric_only=True)

sns.heatmap(data = correlation, annot=True, fmt='.1f',cmap='magma',linewidths=2,square=False)
```

### RUGPLOT

```python
sns.rugplot(data=penguins, x='body_mass_g', hue='sex', height=0.05)
sns.kdeplot(data=penguins, x='body_mass_g',color='#ffb41f',fill=True)
```

### PAIR PLOT

```python
sns.pairplot(data=penguins, hue='species',diag_kind='kde',corner=True)
```

### PAIR GRID

```python
pairGrid = sns.PairGrid(data=penguins, hue='species')

pairGrid.map_upper(sns.scatterplot)
pairGrid.map_diag(sns.histplot)
pairGrid.map_lower(sns.kdeplot)
```
